```python
import math
import numpy as np
```

```python
def lr(o, e):
    return 0.4 + 0.1 * math.exp(2 * (o - e))
```

```python
for i in np.linspace(1, 7, 60):
    print(i, lr(i, 2.42))
```

```python
def obs(lr, e):
    return e + math.log((lr - 0.4) / 0.1) / 2
```

```python
for i in [0.41, 0.44, 0.72, 2.75, 17.78]:
    print(i, obs(i, 2.42))
```

```python
for i in np.linspace(1, 7, 60):
    print(i, lr(i, 4.98))
```

```python
def nuchal(bpd):
    return (0.131 + 0.08 * bpd, 1.219 + 0.08 * bpd)
```

```python
for i in np.arange(28, 62, 2):
    print(i, nuchal(i))
```

```python
def likelyhood_ratio(bpd, nuchal):
    expected = 0.131 + 0.08 * bpd
    return 0.4 + 0.1 * math.exp(2 * (nuchal - expected))
```

```python
likelyhood_ratio(41.908, 2.232562)
```
